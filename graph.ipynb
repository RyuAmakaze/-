```python
import glob
import numpy as np
import matplotlib.pyplot as plt

files = glob.glob("all_lap/*.csv")
for file in files:  
    print(file)
    data_set = np.loadtxt(
    fname=file, #読み込むファイルのパスと名前
    dtype="float", #floatで読み込む
    delimiter=",", #csvなのでカンマで区切る
    encoding='utf-8',
    skiprows = 1
    )
    
    for data in data_set:
        plt.scatter(data[0], data[1])
    
    plt.title(file) #タイトル
    plt.xlabel("z") #x軸のラベル
    plt.ylabel("Max Grip_Laplacian Variance") #y軸のラベル
    plt.grid() #グリッド線を引く(引かなくてもいい別に)

    plt.show()
    
   
```

```python
files = glob.glob("all_lap/*.csv")
for file in files:
    #print(file)
    data_set = np.loadtxt(
    fname=file, #読み込むファイルのパスと名前
    dtype="float", #floatで読み込む
    delimiter=",", #csvなのでカンマで区切る
    encoding='utf-8',
    skiprows = 1
    )
    z = np.array([])
    max_lap = np.array([])

    for data in data_set:
        z = np.append(z,(data[0]))
        max_lap = np.append(max_lap,data[1])
    print(file+"真の最大値:"+str(max_lap[np.argmax(max_lap)])+" (z="+str(z[np.argmax(max_lap)])+")")
```
