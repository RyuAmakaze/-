```python
import glob
import numpy as np
import matplotlib.pyplot as plt
import math

files = glob.glob("*.csv")
for file in files:  
    print(file)
    data_set = np.loadtxt(
    fname=file, #読み込むファイルのパスと名前
    dtype="float", #floatで読み込む
    delimiter=",", #csvなのでカンマで区切る
    encoding='utf-8',
    skiprows = 1
    )
    
    z = 0
    y = 0
    for data in data_set:
        plt.scatter(data[0], (y+data[1])/math.sqrt(data[1]*y))
        y = data[1]
        
    plt.title(file) #タイトル
    plt.xlabel("z") #x軸のラベル
    plt.ylabel("Max Grip_Laplacian Variance") #y軸のラベル
    plt.grid() #グリッド線を引く(引かなくてもいい別に)

    plt.show()
```
